```python
#install packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read html
urlString='https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yield '
dfs = pd.read_html(urlString)
```

```python
#return the number of tables
len(dfs)
```

```python
#find the table of interest rate
dfs[0]
```

```python
dfs[1]
```

```python
#Set table and index
df_yield = dfs[1]
df_yield.set_index('Date',inplace=True)
```

```python
df_yield
```

```python
#extract the latest yields
row = df_yield.iloc[-1:]
row
```

```python
#Plot the latest yield curve.
y = row
x = [1/12, 2/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30]
plt.scatter(x, y)
plt.show()
```

```python
#Plot the time series of the 2 year and 10 year
fig = plt.figure(figsize=(12,3))
ax1, ax2 = fig.subplots(1, 2)
ax1.plot(df_yield['10 yr'])
ax2.plot(df_yield['2 yr'])
plt.legend() 
plt.show() 
```
